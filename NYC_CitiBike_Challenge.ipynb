```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
bikeshare_df = pd.read_csv('201908-citibike-tripdata.csv')
```

```python
# 2. Check the datatypes of your columns. 
bikeshare_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
bikeshare_df['tripduration_sec']=bikeshare_df['tripduration']
bikeshare_df['tripduration']= pd.to_datetime(bikeshare_df['tripduration'], unit='s')
```

```python
# Transform Gender as string object
def genderString(gender):
    if (gender==1):
        return 'MALE'
    elif (gender==2):
        return 'FEMALE'
    return 'UNKNOWN'
bikeshare_df['gender_str'] = bikeshare_df['gender'].map(genderString)
```
