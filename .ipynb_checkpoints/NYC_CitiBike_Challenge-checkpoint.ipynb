```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
bikeshare_df = pd.read_csv('201908-citibike-tripdata.csv')
```

```python
# 2. Check the datatypes of your columns. 
bikeshare_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
bikeshare_df['tripduration_sec']=bikeshare_df['tripduration']
bikeshare_df['tripduration']= pd.to_datetime(bikeshare_df['tripduration'], unit='s')
```

```python
# Transform Gender as string object
if bikeshare_df['gender']== 2:
    bikeshare_df['gender_str']='FEMALE'
elif bikeshare_df['gender']== 1:
    bikeshare_df['gender_str']='MALE'
else:
    bikeshare_df['gender_str']='UNKNOWN'
```

```python
# 4. Check the datatypes of your columns. 
bikeshare_df.dtypes
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
bikeshare_df.to_csv('bikeshare-tripdata-edited.csv')
```
